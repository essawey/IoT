# Preprocessing

````python
# Create Resampling Function with new space = [0.5,0.5,0.5]
## Add your Code Here
def resample_img(img, target_voxel_spacing):
  from scipy.ndimage import zoom

  # Get the image data and header
  data = img.get_fdata()
  header = img.header

  # Get the current voxel spacing and image shape
  current_voxel_spacing = header.get_zooms()
  current_image_shape = data.shape

  # Calculate the resampling factor
  resampling_factor = np.divide(current_voxel_spacing, target_voxel_spacing)

  # Calculate the target image shape
  target_image_shape = np.round(np.multiply(current_image_shape, resampling_factor)).astype(int)

  # Resample the image
  resampled_data = zoom(data, resampling_factor, order=3)

  # Update the header with the new voxel spacing
  header.set_zooms(target_voxel_spacing)

  # Create a new nibabel image object with the resampled data and updated header
  return nib.Nifti1Image(resampled_data, img.affine, header)
````

````python
# mni_img = nib.load(tflow.get('MNI152NLin2009cAsym', resolution=1, suffix="T1w", desc=None))
# t1_img = nib.load("ds001233/sub-17/ses-pre/anat/sub-17_ses-pre_T1w.nii.gz")

target_voxel_spacing = (0.5, 0.5, 0.5)

# Volume No.1
resampled_mni_img = resample_img(
                                  img = mni_img,
                                  target_voxel_spacing = target_voxel_spacing,
                                )

# Volume No.2
resampled_t1_img = resample_img(
                                  img = t1_img,
                                  target_voxel_spacing = target_voxel_spacing,
                               )
````

````python
# mni_img = nib.load(tflow.get('MNI152NLin2009cAsym', resolution=1, suffix="T1w", desc=None))
# t1_img = nib.load("ds001233/sub-17/ses-pre/anat/sub-17_ses-pre_T1w.nii.gz")
````

````python
## Resample Volume 2 (t1_img) to Volume 1 (mni_img) --------> That at the end Both have the same Affine Transformation of mni_img
## Print the shape of ti_img that will have the same shape of mni_img after resampling

## Add your Code Here
from nibabel.processing import resample_from_to
t1_img_resample_to_mni_img = resample_from_to(t1_img, mni_img)
````

````python
print(f't1 img shape:                      {t1_img.shape}')
print(f'mni img shape:                     {mni_img.shape}')
print(f't1 img resample to mni img shape:  {t1_img_resample_to_mni_img.shape}')
````

````python
## Display the output(t1_img Resampled and before t1_img) and mni_img ---> Slice [:,:,85]
## Add your Code Here

mni_img_data = mni_img.get_fdata()
t1_img_data = t1_img.get_fdata()
t1_img_resample_to_mni_img_data = t1_img_resample_to_mni_img.get_fdata()


mni_img_slice =  mni_img_data[:, :, 85]
before_t1_img_slice = t1_img_data[:, :, 85]
t1_img_Resampled_slice = t1_img_resample_to_mni_img_data[:, :, 85]

# Display axial view
plt.subplot(131)
plt.title("Mni image")
plt.imshow(
    mni_img_slice,
    cmap='gray'
    )

# Display coronal view
plt.subplot(132)
plt.title("T1 image Before ")
plt.imshow(
    before_t1_img_slice,
    cmap='gray'
    )

# Display sagittal view
plt.subplot(133)
plt.title("T1 image After Resampling")
plt.imshow(
    t1_img_Resampled_slice,
    cmap='gray'
    )


# Show the plot
plt.tight_layout()
plt.show()
````

````python
## Make Registration that will Align features between two volume
## Display (Mni_image,t1_resampled_image Before, t1_resampled_image After registration) ----> [:,:,85]

## Add your Code Here
from dipy.align.transforms import AffineTransform3D
from dipy.align.imaffine import AffineRegistration

AffineRegistration = AffineRegistration()

affine_map = AffineRegistration.optimize(static = mni_img.get_fdata(),
                                         moving = t1_img.get_fdata(),
                                         transform = AffineTransform3D(),
                                         params0 = None,
                                         static_grid2world = mni_img.affine,
                                         moving_grid2world = t1_img.affine
                                         )
````

````python
## Expected Output
````

````python
t1_before_registration_img = t1_img_resample_to_mni_img_data

mni_img_data = mni_img.get_fdata()
t1_img_resample_to_mni_img_data = t1_img_resample_to_mni_img.get_fdata()
t1_after_registration_img_data = affine_map.transform(t1_img.get_fdata())
````

````python
## Display the output(t1_img Resampled and before t1_img) and mni_img ---> Slice [:,:,85]
## Add your Code Here

mni_img_data_slice =  mni_img_data[:, :, 85]
t1_before_registration_img_data_slice = t1_img_resample_to_mni_img_data[:, :, 85]
t1_after_registration_img_data_slice = t1_after_registration_img_data[:, :, 85]

plt.figure(figsize=(9, 3))

# Display axial view
plt.subplot(131)
plt.title("Mni image")
plt.imshow(
    mni_img_data_slice,
    cmap='gray'
    )

# Display coronal view
plt.subplot(132)
plt.title("T1 image Before Registration")
plt.imshow(
    t1_before_registration_img_data_slice,
    cmap='gray'
    )

# Display sagittal view
plt.subplot(133)
plt.title("T1 image After Registration")
plt.imshow(
    t1_after_registration_img_data_slice,
    cmap='gray'
    )

# Show the plot
plt.tight_layout()
plt.show()
````
